```python
import os 
import json 
from matplotlib import pyplot as plt 
```

```python
files = [i for i in os.listdir() if i.endswith('.json') and not i.startswith('Final')]
files
```

```python
for file_name in files:
    temp_dict = dict()
    with open(file_name,'r') as json_file:
        temp_dict = json.load(json_file)
    plt.close()
    plt.plot([i+1 for i in range(len(temp_dict))],list(temp_dict.values()),'-o')
    plt.grid(True)
    plt.xlabel('Iterations')
    plt.ylabel('Accuracy')
    plt.title(f'{file_name.split("_")[0]} Algorithm (NIA)')
    plt.savefig(f'{file_name.split("_")[0]}_100_Iters.svg',format='svg')
    plt.show()
```
